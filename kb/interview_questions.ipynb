# Questions

1) extract all invoices from the database with SQL and Python
2) calculate sales by customer id using SQL and Python
3) find top 10 biggest customers by sales value using pandas dataframe

```python
import sqlite3 as lite
con = lite.connect('chinook.sqlite')
cur = con.cursor()
sql = """SELECT name FROM sqlite_master 
        WHERE 
            type ='table' AND 
            name NOT LIKE 'sqlite_%';
    """

cur.execute(sql)
cur.fetchall()
```

```python
import sqlite3 as lite
con = lite.connect('chinook.sqlite')
cur = con.cursor()
sql = """SELECT * FROM Invoice;
    """

cur.execute(sql)
cur.fetchall()
```

```python
import sqlite3 as lite
import pandas as pd
con = lite.connect('chinook.sqlite')
cur = con.cursor()
sql = """SELECT Invoice.CustomerId, SUM(InvoiceLine.UnitPrice * InvoiceLine.Quantity) AS sales from INVOICE
LEFT JOIN InvoiceLine
ON Invoice.InvoiceId=InvoiceLine.InvoiceId
GROUP BY Invoice.CustomerId;
    """

cur.execute(sql)
cur.fetchall()

df = pd.read_sql(sql,con)
```

```python
df = df.sort_values(by='sales', ascending=False)
df.head(n=10)
```
